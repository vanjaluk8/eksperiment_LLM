```python
import pandas as pd

df1 = pd.read_csv('metrike/prompts/llm_stress_test_results.csv')
df2 = pd.read_csv('metrike/prompts/llm_chat_test_results.csv')
df3 = pd.read_csv('metrike/prompts/llm_summarize_test_results.csv')
df4 = pd.read_csv('metrike/prompts/llm_batch_test_results.csv')

df = pd.concat([df1, df2, df3, df4], ignore_index=True)

df.head(10)
```

```python
## Define the models of interest
models_to_analyze = ['gemma3', 'mistral', 'meta-llama']
test_names_to_analyze = ['stress', 'chat', 'summarizer', 'batch']
# Filter the DataFrame to only include the specified models
df_filtered_models = df[
    (df['model'].isin(models_to_analyze)) &
    (df['test_name'].isin(test_names_to_analyze))
]

# Group by model and calculate the desired statistics for the 'ttft' column
statistics = df_filtered_models.groupby('model')['ttft'].agg(
    median='median',
    p99=lambda x: x.quantile(0.99),
    average='mean',
    std_dev='std',
    min='min',
    max='max',
).reset_index()

statistics

```
