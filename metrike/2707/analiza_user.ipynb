```python
import pandas as pd

from triton_analiza_v2 import combined_df

df1 = pd.read_csv('prompts_single/1_prompt_test_results.csv')
df2 = pd.read_csv('prompts_single/2_chat_test_results.csv')
df3 = pd.read_csv('prompts_single/3_summarize_test_results.csv')
df4 = pd.read_csv('prompts_single/4_batch_test_results.csv')

df = pd.concat([df1, df2, df3, df4], ignore_index=True)

df.head(10)
```

```python
# remove latency outliers
def remove_latency_outliers(df, column='latency'):
    """
    Remove outliers from the specified latency column in the DataFrame.
    Outliers are defined as values greater than 3 standard deviations from the mean.
    """
    mean = df[column].mean()
    std_dev = df[column].std()
    threshold = mean + 3 * std_dev
    return df[df[column] <= threshold]
df = remove_latency_outliers(df, 'latency')
df
```

```python
# calculate median, p99 and average for ttft and tpot and group by test_name

def calculate_statistics(df, group_by_column, target_column):
    """
    Calculate median, p99, average, standard deviation, min, and max for the specified target column,
    grouped by the specified column.
    """
    return df.groupby(group_by_column)[target_column].agg(
        median='median',
        p99=lambda x: x.quantile(0.99),
        average='mean',
        std_dev='std',
        min='min',
        max='max',
    ).reset_index()
```

```python
# use the function to calculate for gemma3
statistics = calculate_statistics(df[df['model'] == 'gemma3'], 'test_name', 'ttft')
statistics
```

```python
# use the function to calculate for gemma3
statistics = calculate_statistics(df[df['model'] == 'mistral'], 'test_name', 'ttft')
statistics
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
user_df['test_type'] = 'client'
locust_df['test_type'] = 'stress'
combined_df = pd.concat([user_df, locust_df], ignore_index=True)
# Pretpostavljamo da već imaš učitane podatke u combined_df
# Ako ne, učitaj CSV-jeve i pripremi combined_df kao ranije

# Filtriraj samo Chat i Batch testove
pareto_data = combined_df[combined_df["test"].isin(["Chat", "Batch"])].copy()

# Pivotiranje da imamo Chat i Batch TPS za svaki model u jednom redu
pareto_pivot = pareto_data.pivot(index="model", columns="test", values="TPS").reset_index()

# Izračunaj razliku TPS (Chat - Batch)
pareto_pivot["TPS_diff"] = pareto_pivot["Chat"] - pareto_pivot["Batch"]

# Sortiranje po razlici
pareto_sorted = pareto_pivot.sort_values(by="TPS_diff", ascending=False).reset_index(drop=True)

# Kumulativni postotak
pareto_sorted["% Kumulativno"] = pareto_sorted["TPS_diff"].cumsum() / pareto_sorted["TPS_diff"].sum() * 100

# --- PARETO GRAF ---
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar graf (razlike)
sns.barplot(data=pareto_sorted, x="model", y="TPS_diff", ax=ax1, color="skyblue")
ax1.set_ylabel("Razlika TPS (Chat - Batch)", color="blue")
ax1.set_xlabel("Model")
ax1.tick_params(axis='y', labelcolor="blue")

# Crta (kumulativni %)
ax2 = ax1.twinx()
ax2.plot(pareto_sorted["model"], pareto_sorted["% Kumulativno"], color="red", marker="o", linestyle="-")
ax2.set_ylabel("Kumulativni %", color="red")
ax2.tick_params(axis='y', labelcolor="red")
ax2.set_ylim(0, 110)

plt.title("Pareto graf: razlika TPS (Chat - Batch) po modelu")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
